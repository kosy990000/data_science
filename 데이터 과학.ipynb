```python
#network
#edge -> [start, next]

def pagerank(edges, n_nodes, beta=0.85, epoch=100):
  r = [1/n_nodes]* n_nodes
  d = [0]* n_nodes

  for u, _ in edges:
    d[u] += 1

  teleport = (1-beta) /n_nodes

  for _ in range(epoch):
    r_next = [teleport] * n_nodes

     #시작 노드의 점수를 끝 노드 점수에 합산
    for u, v in edges:
      r_next[v] += beta * r[u]/d[u]


    r_sum = sum(r_next)

    #정규화 dead end 처리
    for u in range(n_nodes):
      r_next[u] += (1-r_sum)/n_nodes

    r = r_next

  return r
```

```python
import networkx as nx
# 방향이 없는 그래프 생성
G = nx.Graph()

# 노드 추가
#G.add_node(1)  # 노드 1 추가
G.add_nodes_from(["A", "B", "C", "D", "E", "F","G", "H" ])  # 여러 노드 추가

# 엣지 추가
#G.add_edge(1, 2)  # 1 -> 2
G.add_edges_from([("A", "B"), ("B", "C"), ("A", "D"), ("C", "D"), ("E", "D"), ("E", "H"),
                  ("G", "H"), ("H", "F"), ("B", "D")])  # 여러 엣지 추가

print("Nodes:", G.nodes)
print("Edges:", G.edges)
```

```python
node_ = len(G.nodes)
edges = len(G.edges)
n_triangles = sum(nx.triangles(G.to_undirected()).values()) //3

density = edges*2/(node_*(node_-1))
print(n_triangles)
print(density)

degree = nx.degree_centrality(G)
#indegree = nx.in_degree_centrality(G)
#outdegree = nx.out_degree_centrality(G)
betweenness = nx.betweenness_centrality(G) #bewteen 계산량 제한
closeness = nx.closeness_centrality(G)
pagerank = nx.pagerank(G)
clustering = nx.clustering(G)
triangles = nx.triangles(G.to_undirected())
diameter = nx.diameter(G)

print(degree)
print(closeness)
print(pagerank)
print(clustering)
print(n_triangles)
print(diameter)

#effective diameter 계산
shortest_paths = dict(nx.all_pairs_shortest_path_length(G))
print(shortest_paths)


list_a = []
for node, distances in shortest_paths.items():
    max_value = 0  # 최대값 초기화
    for neighbor, distance in distances.items():
        if max_value < distance:
            max_value = distance
    list_a.append(max_value)



i = len(list_a) *0.9
i = round(i-1)


sorted_list = sorted(list_a)
print(sorted_list)
print(sorted_list[i])
```
